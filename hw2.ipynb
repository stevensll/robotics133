```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

```python
#QUESTION 3
def theta1(w,t):
    return w*t
def theta2(w,t):
    return w*t + np.pi
def theta3(w,t):
    return w*t + np.pi
l1 = 6
l2 = 1
l3 = 0.5
w1 = 1
w2 = -4
w3 = 12
time_range = np.linspace(-999, 999, 10000)
x = [(l1*np.cos(theta1(w1,t))+
     l2*np.cos(theta1(w1,t)+theta2(w2,t))+
     l3*np.cos(theta1(w1,t)+theta2(w2,t)+theta3(w3,t)))
     if (l1*np.cos(theta1(w1,t))+
     l2*np.cos(theta1(w1,t)+theta2(w2,t))+
     l3*np.cos(theta1(w1,t)+theta2(w2,t)+theta3(w3,t))) 
     is not None else None
     for t in time_range]
y = [(l1*np.sin(theta1(w1,t))+
     l2*np.sin(theta1(w1,t)+theta2(w2,t))+
     l3*np.sin(theta1(w1,t)+theta2(w2,t)+theta3(w3,t)))
     if (l1*np.sin(theta1(w1,t))+
     l2*np.sin(theta1(w1,t)+theta2(w2,t))+
     l3*np.sin(theta1(w1,t)+theta2(w2,t)+theta3(w3,t)))
     is not None else None
     for t in time_range]

plt.plot(x,y)
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.xlim(-(l1+l2+l3),l1+l2+l3)
plt.ylim(-(l1+l2+l3),l1+l2+l3)
plt.title(f'Path of 3 arm robot')
plt.show()
```

```python
#QUESTION 4
d = 2
def theta_pan(t):
    return np.pi/3*np.sin(2*t)

def theta_tilt(t):
    return np.pi/3*np.sin(t) - np.pi/9*np.cos(6*t)

time_range = np.linspace(-999,999,100000)
x = [d*np.tan(theta_pan(t)) if np.tan(theta_pan(t)) is not None else None for t in time_range]
y = [d*np.tan(theta_tilt(t)) if np.tan(theta_tilt(t)) is not None else None for t in time_range]

plt.plot(x,y)

plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title(f'Picture Gimbo Draws on Sreen')
plt.show()
```

```python
#QUESTION 5
l1=1
l2=1
x = 0.05
def theta1(t, theta):
    return (np.arctan2(np.cos(t),x)-
            np.arctan2(np.sin(theta), 1+np.cos(theta)))
def theta2(t):
    return -np.arccos((np.cos(t)*np.cos(t)-l1-l2+x)/2)

time_range = np.linspace(0, 2*np.pi, 10000)
theta1_range = [theta1(t,theta2(t)) for t in time_range]
theta2_range = [theta2(t) for t in time_range]

plt.xlabel("time")
plt.ylabel("Angle (radians)")
plt.plot(time_range,theta1_range, label ="θ₁")
plt.plot(time_range,theta2_range, label ="θ₂")
plt.title("Angle versus time")
plt.legend()
plt.show
```
